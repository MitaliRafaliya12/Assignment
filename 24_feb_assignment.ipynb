Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
function should print the sum to the console.

```python
df = pandas.DataFrame({'A':[1,2,3,56,43,23,12],
                 'Values':[4,5,6,7,6,5,4] , 
                 'C':[7,8,9,7,8,65,43]})
```

Q5. How are DataFrame.size() and DataFrame.shape() different?

DataFrame.size returns the total number of elements in a DataFrame. It is equivalent to the number of rows multiplied by the number of columns.

DataFrame.shape returns a tuple representing the dimensions of the DataFrame. The tuple contains two values: the number of rows and the number of columns, respectively.

```python
df = pandas.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

# Print the size of the DataFrame
print(df.size)  # Output: 9

# Print the shape of the DataFrame
print(df.shape)  # Output: (3, 3)
```
